```python
# import json files
import json
import csv
JSON_DIR = "../data/Winter2025/json/"
EXPORT_DIR = "../data/Winter2025/filtered/"
JSON_FILE = "monthly-statement-2025-04-07.json"
EXPORT_FILE = "monthly-statement-2025-04-07.csv"
RESULT_DIR = "../results/"
RESULT_FILE = "Winter2025_compute_costs.csv"
# load json file
with open(JSON_DIR + JSON_FILE, 'r') as file:
    data = json.load(file)
```

```python
# print data
print(data)

# print data keys
print(data[0].keys())

# filter by id "4721d6f7-2c22-4a6c-9227-499a9b0a40b0"
filtered_data = [item for item in data if item['project_id'] != '7be9898f-f0d6-49af-8532-98825877a784']
filtered_data = [item for item in filtered_data if 'service' in item.keys()]
# filter by service "compute-server"
for item in filtered_data:
    if 'service' in item.keys():
        if item['service'] == "compute-server":
            item['machineType'] = item['description']['configuration']['machineType']
            item['compute_server_id'] = item['description']['compute_server_id']
            item['diskSizeGb'] = item['description']['configuration']['diskSizeGb']
        else:
            item['machineType'] = None
            item['compute_server_id'] = None

# remove elements that are not compute-server
compute_server_data = [item for item in filtered_data if item['service'] == "compute-server"]

del filtered_data

# colums: id, time, cost, period_start, period_end, cost_per_hour, invoice_id, project_id, machineType, compute_server_id
# rows: each item in compute_server_data
#keys_to_include = ['id', 'time', 'cost', 'period_start', 'period_end', 'cost_per_hour', 'invoice_id', 'project_id', 'service','machineType', 'compute_server_id']


# write the filtered data to the csv file
with open(EXPORT_DIR + EXPORT_FILE, 'w') as file:
    writer = csv.writer(file)
    writer.writerow(compute_server_data[0].keys())
    for item in compute_server_data:
        writer.writerow(item.values())
```

```python
# concatenate  all csv in EXPORT_DIR
import os
# create a new csv file with header
with open(RESULT_DIR + RESULT_FILE, 'w', newline='') as result_file:
    writer = csv.writer(result_file)
    writer.writerow(compute_server_data[0].keys())
    
    # read and write all other files
    for filename in os.listdir(EXPORT_DIR):
        if filename.endswith(".csv"):
            try:
                with open(EXPORT_DIR + filename, 'r') as csv_file:
                    reader = csv.reader(csv_file)
                    next(reader, None)  # skip header safely
                    for row in reader:
                        writer.writerow(row)
            except Exception as e:
                print(f"Error processing {filename}: {str(e)}")

```
